```python
import torch
```

```python
# Constract a tensor require grad
x = torch.ones(2, 2, requires_grad = True)
print(x)
```

```python
y = x + 2
print(y)
```

```python
# y is generated by result of manipulation, so it has "grad_fn"
print(y.grad_fn)
```

```python
z = y * y * 3
out = z.mean()
# Manipulation on tensor will change flag grad_fn with tensor has require_grad is True
print(z, out)
```

```python
a = torch.randn(2, 2)
a = ((a * 3) / (a - 1))
print(a.requires_grad)
a.requires_grad_(True)
print(a.requires_grad)
b = (a * a).sum()
print(b.grad_fn)
```

```python
out.backward()
print(x.grad)
```

```python
x = torch.randn(3, requires_grad = True)
y = x * 2;
while y.data.norm() < 1000:
    y = y * 2
print(y)
```

```python
v = torch.tensor([0.1, 1.0, 0.0001], dtype = torch.float)
y.backward(v)
print(x.grad)
```

```python
print(x.requires_grad)
print((x ** 2).requires_grad)

# use with 
with torch.no_grad():
    print((x ** 2).requires_grad)
```
